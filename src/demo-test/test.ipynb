```python
from core import RSNA_IAD
%matplotlib inline 
from matplotlib import pyplot as plt
```

```python

engine = RSNA_IAD(
    exp0_aneurysm_det_ckpt='../exp0_aneurysm_det/checkpoints/yolo11x_1280_fold0/weights/best.pt', 
    exp1_brain_det_ckpt='../exp1_brain_det_yolov5_7.0/checkpoints/yolov5n_640/weights/brain_det_yolov5n_640.pt', 
    exp2_vit_large_ckpt='../exp2_cls/checkpoints/vit_large_patch14_clip_336.openai_ft_in12k_in1k_384_full_epoch4_exp2.pt',
    exp2_eva_large_ckpt='../exp2_cls/checkpoints/eva_large_patch14_336.in22k_ft_in22k_in1k_384_full_epoch10_exp2.pt', 
    exp3_mit_b4_ckpt='../exp3_aux/checkpoints/mit_b4_384_fold1.pt',
    exp4_vit_large_ckpt='../exp4_cls_pseudo/checkpoints/vit_large_patch14_clip_336.openai_ft_in12k_in1k_384_full_epoch4_exp4.pt',
    exp4_eva_large_ckpt='../exp4_cls_pseudo/checkpoints/eva_large_patch14_336.in22k_ft_in22k_in1k_384_full_epoch10_exp4.pt',
    exp5_mit_b4_ckpt='../exp5_aux_pseudo/checkpoints/mit_b4_384_fold2.pt',
    device="cuda:0"
)
```

```python
ser_preds, localizer = engine.predict(
    series_path='../../dataset/kaggle_evaluation/series/1.2.826.0.1.3680043.8.498.10028406715369553772267826812576760572',
    batch_size=8,
    crop_rat=0.75,
    aneurysm_thres=0.5, 
    aneurysm_det_thres=0.2
)
print('*'*10, 'series prediction', '*'*10)
for class_name in engine.classes:
    print('{:50s}: {:.5f}'.format(class_name, ser_preds[class_name]))
if localizer is not None:
    print('*'*10, 'localization', '*'*10)
    print('SOPInstanceUID:',localizer['SOPInstanceUID'])
    fig = plt.figure(figsize=(7, 7))
    ax = fig.add_axes([0, 0, 1, 1])
    ax.axis('off')
    ax.imshow(localizer['image'])
    plt.show()
```
